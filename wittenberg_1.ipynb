# networkx

```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

Let's create some graphs:

```python
G=nx.Graph()
G.add_nodes_from(['Goettingen', 'Hanover', 'Berlin', 'Wittenberg'])
```

```python
G.nodes()
```

```python
G.edges()
```

```python
G.add_edges_from([('Goettingen', 'Hanover'), ('Goettingen', 'Berlin'), ('Hanover','Berlin'), ('Berlin', 'Wittenberg'), ('Goettingen', 'Leipzig'), ('Leipzig', 'Wittenberg'), ('Berlin', 'Leipzig')])
```

```python
G.edges()
```

```python
nx.draw(G)
```

```python
nx.draw_networkx(G, node_size=500, alpha=0.5, width=2, node_color='green')
```

```python
for path in nx.all_shortest_paths(G, source='Goettingen', target='Wittenberg'):
    print path
```

### networkx graphs are built on top of dictionaries

Another cool thing is that, edges are all *empty dictionaries*: meaning any key-value pairs can be stored in them:

```python
G['Berlin']['Hanover']['distance']=1.8
G['Goettingen']['Berlin']['distance']=2.0
G['Berlin']['Leipzig']['distance']=1
G['Berlin']['Wittenberg']['distance']=0.7
G['Leipzig']['Wittenberg']['distance']=0.4
G['Hanover']['Goettingen']['distance']=0.5
G['Goettingen']['Leipzig']['distance']=4
```

```python
G[u'Berlin'][u'Hanover']
```

```python
for path in nx.all_shortest_paths(G, source='Goettingen', target='Wittenberg', weight='distance'):
    print path
```

```python
nx.shortest_path_length(G, source='Goettingen', target='Wittenberg', weight='distance')
```

*nodes can also hold attributes:*

```python
G.node[u'Berlin']['population']=14
```

```python
G.node[u'Berlin']
```
