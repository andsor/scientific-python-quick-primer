```python
import IPython.html.widgets as widgets
```

```python
def sincos(a, b):

    fig, axes = plt.subplots()
    axes.plot(x, a * np.sin(x + b), label="$a \sin(x) + b$")
    axes.plot(x, np.cos(x), color="red", marker="o", linestyle="--", label="$\cos(x)$")
    axes.set_xlabel("$x$", size=20)
    axes.set_ylabel("$\sin(x), \cos(x)$", size=20)
    axes.legend(); 
    
    axes.set_xlim(-2 * np.pi, 2 * np.pi)
    axes.set_ylim(-2, 2)
```

# Examples

## The Logistic Map

```python
def iterate_logistic_map(r, x):
    return r * x * (1. - x)

r = np.linspace(3, 4, num=1000)
x = np.random.rand(1000)
r_mesh, x_mesh = np.meshgrid(r, x)

for i in range(1000):
    x_mesh = iterate_logistic_map(r_mesh, x_mesh)
```

```python
fig, axes = plt.subplots()
axes.plot(r_mesh, x_mesh, color='red', ls='', marker='.', markersize=0.25, rasterized=True)
axes.set_xlim(r.min(), r.max())
axes.set_xlabel('$r$')
axes.set_ylabel('$x$');
```

## The Random Walk

```python
n_steps = 1000
n_samples = 100000

steps = 2 * np.random.randint(0, 2, (n_steps, n_samples)) - 1
trajectories = steps.cumsum(axis=0)

# Plot 7 trajectories
fig, axes = plt.subplots()
axes.plot(trajectories[:, :7], rasterized=True)
axes.set_xlabel('$n$')
axes.set_ylabel('$S_n$');
```

```python
import scipy.stats

bound = max(np.abs(trajectories.max()), np.abs(trajectories.min()))

hist = np.histogram(trajectories[-1, :], bins=np.arange(-bound, bound + 1, 2), normed=True, density=True)

fig, axes = plt.subplots()
axes.step(hist[1][:-1] + 1, hist[0])
axes.plot(hist[1], scipy.stats.norm.pdf(hist[1], scale=np.sqrt(1000)), lw=2)
axes.set_xlabel('$S_{{{}}}$'.format(n_steps));
```

### Bonus: interactive histogram

```python
hists = np.zeros((n_steps, bound))
for step in range(1, n_steps):
    hists[step, :] = np.histogram(trajectories[step, :], bins=np.arange(-bound, bound + 1, 2), normed=True, density=True)[0]
    
def plot_histogram(step):
    fig, axes = plt.subplots()
    axes.step(hist[1][:-1] + 1, hists[step, :], lw=2)
    axes.plot(hist[1], scipy.stats.norm.pdf(hist[1], scale=np.sqrt(step)), lw=3, alpha=0.66);
    axes.set_ylim(0, 0.25)
```

```python
import IPython.html.widgets as widgets
widgets.interact(plot_histogram, step=(1, 999, 1))
```

## The Kuramoto Model

```python
from scipy.integrate import odeint
import scipy.stats

def f(y, _, omega, K, N):
    yk, yi = np.meshgrid(y, y)
    return omega + K / N * np.sum(np.sin(yk - yi), axis=1)

def run(N, scale, K):
    omega = scipy.stats.norm.rvs(loc=0.0, scale=scale, size=N)
    omega -= omega.mean()
    y0 = np.random.uniform(0.0, 2 * np.pi, size=N)
    t1 = 100
    dt = t1 / 400
    
    trajectories = odeint(f, y0, np.arange(t1, step=dt), args=(omega, K, N))
    order = np.abs(np.mean(np.exp(trajectories * 1j), axis=1))
    
    return trajectories, order
```

```python
fig, axes = plt.subplots(nrows=3)
for i, K in enumerate([0.1, 0.24, 10.0]):
    trajectories, order = run(N=6, scale=0.1, K=K)
    axes[i].set_ylim(0, 2 * np.pi)
    axes[i].plot(
        np.mod(trajectories, 2 * np.pi),
        ls='', marker='.', markersize=2
    )
    ax2 = axes[i].twinx()
    ax2.plot(order, color='black', linewidth=2)
    ax2.set_ylim(0, 1)
    if i == 1:
        ax2.set_ylabel('$r$')
    
axes[-1].set_xlabel('$t$')
axes[1].set_ylabel(r'$\theta$');
```
